```python
#29.Remove the CarID column from the dataset.
df.drop(columns=['CarID'], inplace=True)
```

```python
#30.	Reset the index of the DataFrame.
df.reset_index(drop = True, inplace = True)
```

```python
#31.	Replace all unsold cars ("No") with "Available" in the Sold column.
df['Sold'] = df['Sold'].replace({'No':'Available'})
```

```python
df.head()
```

```python
#32.Create a new column PerformanceCategory based on Horsepower (High: >300, Medium: 150-300, Low: <150).

df['PerformanceCategory'] = None
conditions = [
    (df['Horsepower'] > 300),
    (df['Horsepower'] >= 150) & (df['Horsepower'] <= 300),
    (df['Horsepower'] < 150)
]

choices = ['High', 'Medium', 'Low']

df['PerformanceCategory'] = np.select(conditions, choices, default='Unknown')

print(df[['Horsepower', 'PerformanceCategory']])


```

```python
df.head()
```

```python
#33.	Convert the Year column to an integer type.

df['Year'].astype('int64')
```

```python
#34.	Create a new column AboveAveragePrice, which is "Yes" if Price is above the average.
df['Price'].mean()
df['AboveAveragePrice'] = (df['Price'] > df['Price'].mean()).map({True: 'Yes', False: 'No'})

```

```python
#35.	Convert the FuelType column to a categorical type.
df['FuelType'] = df['FuelType'].astype('category')
```

```python
#36.Find the top 5 most expensive cars.
df.nlargest(5,'Price')
```

```python
#38.	Save the modified DataFrame to a new CSV file.
df.to_csv('modified_data.csv', index=False)
```
