5.	Write a Python program to convert the data type of a NumPy array from float to int using astype().
o	Array = [5.5, 6.7, 8.9]


```python
import numpy as np
a = np.array([5.5, 6.7, 8.9])
b = a.astype('int64')
print(b)
```
