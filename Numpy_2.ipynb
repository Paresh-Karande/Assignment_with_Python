Given a 2D NumPy array arr_2d, transpose the array and print the result.

Example Input:


arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

Example Output:


Transposed Array:
[[1 4]
 [2 5]
 [3 6]]



```python
import numpy as np
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
b = np.transpose(arr_2d)
print(b)
```
